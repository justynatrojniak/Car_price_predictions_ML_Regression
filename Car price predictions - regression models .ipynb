```python
import pandas as pd
```

```python
# loading data from csv file, (with header = None pandas set the header by an integer from 0)

data = pd.read_csv('/Users/apple/Downloads/auto.csv', header=None)
```

```python
# 5 first rows of our data
data.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#replacing headers with correct names 

data.columns = headers
```

```python
# replacing ? with NaN for reason of computational speed and convenience 

import numpy as np
data.replace('?', np.nan, inplace = True)
```

```python
data.head()
```

```python
# evaluating for missing data, True stands for missing data
missing_data = data.isnull()
missing_data.head()
```

```python
# detecting and counting missing values in each column

for x in missing_data.columns.values.tolist():
    print(x)
    print(missing_data[x].value_counts(), '\n')
```

```python
# each column has 205 rows, hence 7 columns contain missing data: 
# normalized-losses 41, bore 4, stroke 4, horsepower 2, peak-rpm 2 - will be replaced by mean
# num-of-doors 2 - will be replaced by frequency
# price 4 - rows with missing data will be dropped as price is what we want to predict therefore any rows without price is useful


avg_norm_loss = data['normalized-losses'].astype('float').mean(axis=0)
avg_norm_loss
```

```python
# replacing NaN with means

data['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
avg_bore = data['bore'].astype('float').mean(axis=0)
data['bore'].replace(np.nan, avg_bore, inplace = True)
```

```python
print(avg_bore)
```

```python
avg_stroke = data['stroke'].astype('float').mean(axis=0)
data['stroke'].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_horsepower = data['horsepower'].astype('float').mean(axis=0)
data['horsepower'].replace(np.nan, avg_horsepower, inplace = True)
```

```python
avg_peak = data['peak-rpm'].astype('float').mean(axis=0)
data['peak-rpm'].replace(np.nan, avg_peak, inplace = True)
```

```python
# frequency for num-of-doors - most common type

data['num-of-doors'].value_counts().idxmax()
```

```python
data['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
# dropping nan rows from price

data.dropna(subset=['price'], axis = 0, inplace = True)
```

```python
data.head()
```

```python
# changing data type for numerical values such as bore, stroke, price, peak-rpm, normalized-losses 
# as there is no not numerical values already

data[['normalized-losses', 'horsepower']] = data[['normalized-losses','horsepower']].astype('int')
data[['bore','stroke','price','peak-rpm']] = data[['bore','stroke','price','peak-rpm']].astype('float')
```

```python
data.dtypes
```

```python
# normalization of length, width and height (transforming to range 0-1)

data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
data['height'] = data['height']/data['height'].max()
```

```python
# based on the above analysis, the variables take should be taken into account can be narrowed down to:
# numerical: wheel-base, length, width, curb-weight, engine-size, bore, horsepower, city- and highway- mpg
# categorical: drive-wheels

# model development
# linear regression - when we want to predict the price using one variable

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = data['highway-mpg']
X.dtype
```

```python
# linear regression model
lm = LinearRegression()

# defining variables
X = data[['highway-mpg']] # must be dataframe, not series -> otherwise 'expected 2D array, got 1D array instead' error
Y = data['price']

# splitting into test and train data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = False)

lm.fit(X_train,Y_train)

#predicting the test set results
Yhat = lm.predict(X_test)
Yhat[0:5]
```

```python
# R^2 (R-squared) and MSE (Mean Squared Error)

# R^2 also known as the coefficient of determination indicates how close the data is to the fitted regression line
# MSE measures the difference between actual value and the estimated value

from sklearn.metrics import mean_squared_error

lm.score(X,Y)
print('R^2: ', lm.score(X,Y))

mse = mean_squared_error(Y_test,Yhat)
print(('MSE: ', mse))
```

```python
# visualizing the lr model training results
import matplotlib.pyplot as plt 

plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, lm.predict(X_train), color = 'blue')
plt.title('Price vs Highway-mpg (Training set)')
plt.xlabel('Highway-mpg')
plt.ylabel('Price')
plt.show()
```

```python
# visualizing the lr model testing results

plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, lm.predict(X_test), color = 'green') 
# could be left with training data: X_train, Y_train (the same lines in both cases)
# regression line that we get is actually resulting from a unique equation and therefore the predicted
# salaries of the test set will be on the same regression line as the predicted salaries of the training set

plt.title('Price vs Highway-mpg (Testing set)')
plt.xlabel('Highway-mpg')
plt.ylabel('Price')
plt.show()
```

```python
# residual plot
sns.residplot(X,Y)
plt.show() 
# The residuals are not randomly spread around the axis x 
# that may indicate that a non linear model is more appropriate for this data
```

```python
# multiple linear regression for more than one independent variable
data[['price', 'wheel-base', 'length', 'width', 'curb-weight', 'engine-size', 
      'bore', 'horsepower', 'city-mpg', 'highway-mpg']].corr()
```

```python
# defining Z for more than one independent variable
Z = data[['width','curb-weight', 'engine-size','horsepower', 'highway-mpg']]

lm.fit(Z,Y)
Yhat1 = lm.predict(Z)
Yhat1[0:5]
```

```python
# R^2 and MSE

lm.score(Z, Y)
print('R^2: ', lm.score(Z, Y))

mse1 = mean_squared_error(Y, Yhat1)
print('MSE: ', mse1)
```

```python
ax1 = sns.distplot(data['price'], hist=False, color='r', label = 'Actual Values')
sns.distplot(Yhat1, hist=False, color = 'b', label = 'Fitted Values', ax=ax1)

plt.title('Actual vs Fitted Price Values')
plt.xlabel('Price $')
plt.ylabel('Cars proportion')

plt.show()

# values are reasonably close, but there is some room for improvement
```

```python
# polynomial regression training 

# Z = data[['width','curb-weight', 'engine-size','horsepower', 'highway-mpg']]
# Y = data['price']
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
z_poly = poly.fit_transform(Z)
lin_reg = LinearRegression()
lin_reg.fit(z_poly, Y)
```

```python
poly_predict = lin_reg.predict(z_poly)
poly_predict[0:5] # predicted price for the first five cars
```

```python
# R^2 and MSE

from sklearn.metrics import r2_score

r_squared = r2_score(Y, poly_predict)
print('R^2: ', r_squared)

mse3 = mean_squared_error(Y, poly_predict)
print('MSE: ', mse3)
```

Usually the more variables we have, the better our model is predicting but this is not always true. Sometimes we may not have enough data, we may run into numerical problems or many of the variables may not be useful or even act as a noise. As a result, we should always check the MSE and R^2.

Comparing these 3 models, we can conclude that the MLR model is the best model to be able to predict price from our dataset.
